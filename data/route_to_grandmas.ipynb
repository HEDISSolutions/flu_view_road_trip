```python
import pandas as pd
import os

# Relative path for the input file
input_file = os.path.join("data", "StateDatabyWeek.csv")

# Relative path for the output file
output_folder = "output"
os.makedirs(output_folder, exist_ok=True)  # Ensure the output folder exists
output_file = os.path.join(output_folder, "Route_to_Grandmas.csv")

# List of states to keep (rename the dictionary)
route_to_grandmas = ["California", "Nevada", "New Mexico", "Texas", "Alabama", "Louisiana", "Florida"]

# Load the CSV file
df = pd.read_csv(input_file)

# Filter rows to keep only those where STATENAME is in route_to_grandmas
filtered_df = df[df['STATENAME'].isin(route_to_grandmas)]

# Save the filtered DataFrame to a new CSV file
filtered_df.to_csv(output_file, index=True)


filtered_df

```

```python
# Ensure the 'ACTIVITY LEVEL' column is of type string, remove 'LEVEL', and convert values to integers
filtered_df.loc[:, 'ACTIVITY LEVEL'] = (
    filtered_df['ACTIVITY LEVEL']
    .astype(str)  # Ensure the column is of type string
    .str.replace('LEVEL', '', case=False, regex=False)  # Remove 'LEVEL'
    .str.strip()  # Remove whitespace
    .astype(int)  # Convert cleaned values to integers
)

# Define the path for the new CSV file
new_output_file = os.path.join("data", "route_to_grandmas.csv")

# Save the filtered DataFrame to the new CSV file
filtered_df.to_csv(new_output_file, index=True)

filtered_df
```
